# Importing library modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

#Step Size 0.1

```python
x = np.arange(-10, 10,0.1)
#checking step size values
#print(x)
```

# Plot: y = wx + b

```python
w = 3
b = 5
y = w*x + b

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y)
plt.title('Plot: y=wx+b')
plt.grid('on')
plt.show
```

# Plot: y = X ^ 2

```python
y = np.power(x,2)

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.axhline(0, color='black', lw=3)
plt.axvline(0, color='black', lw=3)

plt.plot(x,y)
plt.title('Plot: x^2')
plt.grid('on')
plt.show
```

# Plot: y = 1/(1+e^(-x))

```python
y = 1/(1+np.exp(-x))

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y)
plt.title('Plot: y=1/(1 + e^(-x))')
plt.grid('on')
plt.show
```

#Plot: y=(e^x - e^(-x))/(e^x-e^(-x))

```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.axhline(0, color='black', lw=3)
plt.axvline(0, color='black', lw=3)

plt.plot(x,y)
plt.title('Plot: y=(e^x - e^(-x)) / (e^x - e^(-x))')
plt.grid('on')
```

#Plot: y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
u = w*x + b
y = 1 / (1 + np.exp(-u))

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y)
plt.title('Plot: y = 1/(1+e^(-x))')
plt.grid('on')
plt.show
```

#Plot: y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
u = w*x + b
y = (np.exp(u) - np.exp(-u)) / (np.exp(u) + np.exp(-u))

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y)
plt.title('Plot: y=(e^x - e^(-x)) / (e^x - e^(-x))')
plt.grid('on')
plt.show
```

#Plot: y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))

```python
w1 = 2
b1 = 8
u1 = w1*x + b1
y1 = 1 / (1 + np.exp(-u1))

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y1)
plt.title('Plot: y = g(f(x)) ,g(u)=1/ (1 + e^(-u))')
plt.grid('on')
plt.show
```

#Plot: y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))

```python
w2 = 4
b2 = 4
u2 = w2*x + b2
y2 = 1 / (1 + np.exp(-u2))

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y2)
plt.title('Plot: y = g(f(x), g(u) = (e^u - e^(-u)) / (e^u + e^(-u))')
plt.grid('on')
plt.show
```

#Plot: y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))

```python
w3 = -12
w4 = -12
b = 15
w = w3*y1 + w4*y2 + b
y = 1 / (1 + np.exp(-w))

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y)
plt.title('Plot: y = g(f(x))')
plt.grid('on')
plt.show
```
